```python
import json
import sys
import cobrakbase
sys.path.append("/Users/chenry/code/ModelSEEDpy")
from modelseedpy.core.eschermapapi import EscherMapAPI, MSEscherMap
from json import JSONEncoder

class MyEncoder(JSONEncoder):
    def default(self, o):
        return o.__dict__
```

```python
kbase_api = cobrakbase.KBaseAPI()
mapapi = EscherMapAPI(kbase_api)

data = {}
with open('/Users/chenry/code/ModelSEEDpy/examples/example_map.json') as json_file:
    data = json.load(json_file)
newmap = MSEscherMap(data[0]["map_id"])
newmap.set_attributes_from_data(data)

newmap.type = "Pathway"

mapapi.save_map(newmap)

maplist = mapapi.list_maps()
print(json.dumps(maplist,cls=MyEncoder,indent=2))
    
savedmap = mapapi.get_maps([newmap.id])
print(json.dumps(savedmap,cls=MyEncoder,indent=2))
```
